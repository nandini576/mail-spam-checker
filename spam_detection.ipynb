```python
import pandas as pd
from sklearn.model_selection import train_test_split;
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import confusion_matrix,accuracy_score, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
data_set=pd.read_csv("C:\\Users\\nandu\\Downloads\\spam.csv", encoding='latin1');
# print(data_set)
data_set = data_set.rename(columns={'v1': 'label', 'v2': 'raw_text'})
data_set=data_set[["label","raw_text"]]
data_set['label']=data_set["label"].map({"ham":0,"spam":1})
# print(data_set)
spam_counts = data_set['label'].value_counts()
labels = ['Ham', 'Spam']
counts=[spam_counts[0],spam_counts[1]]
plt.bar(labels,counts,color=['green','red'],width=0.4)
plt.show()
X=data_set["raw_text"]
Y=data_set["label"]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
# print(X_train.head())
# print(Y_train.head())
# print(X_test.head())
# print(Y_test.head())
vectorizer=CountVectorizer()
X_train_vec=vectorizer.fit_transform(X_train)
X_test_vec=vectorizer.transform(X_test)
# print(X_train_vec)
# print(X_test_vec)
model=LogisticRegression()
model.fit(X_train_vec,Y_train)
predicted_values=model.predict(X_test_vec)
print("Predicted values: ",predicted_values)
print(classification_report(Y_test, predicted_values, target_names=["Ham", "Spam"]))
# cm = confusion_matrix(Y_test, predicted_values)
# sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Ham", "Spam"], yticklabels=["Ham", "Spam"])
# plt.xlabel("Predicted")
# plt.ylabel("Actual")
# plt.title("Confusion Matrix")
# plt.show()

sample_msg=input("enter a message you want to detect as a spam or not")
sample_vec=vectorizer.transform([sample_msg])
prediction=model.predict(sample_vec)
if prediction[0] ==1:
    prediction_msg= 'SPAM'
else:
    prediction_msg= 'NOT SPAM' 
print(f" prediction for \"sample_msg\": is {prediction_msg}")
```
